```python
# ============================================================
# LONG-RUN DES FOR MULTI-CANCER SCREENING AT NYP USING SIMPY
# ============================================================
#
# This file is a starter framework for a patient-level
# discrete-event simulation (DES) using SimPy.
#
# The goal is to model female patients moving through:
#   1. entry into the NYP system
#   2. assignment to one or more screening programs
#   3. screening test selection
#   4. positive / negative screening results
#   5. treatment pathways
#   6. successful treatment / unsuccessful treatment
#   7. repeated re-entry over a long time horizon
#
# IMPORTANT:
# This is a modeling scaffold. It is NOT a clinically validated
# model yet. You and your professor would later replace the
# placeholder probabilities, intervals, and distributions with
# empirical estimates from real NYP data.
# ============================================================


# ---------------------------
# IMPORTS
# ---------------------------

import simpy                      # SimPy provides the DES engine: environment, resources, and events.
import random                     # random is used to generate placeholder probabilities and durations.
import math                       # math is included in case you later want more detailed formulas.
from dataclasses import dataclass # dataclass makes patient records cleaner and easier to manage.
from typing import Dict, List     # These type hints make the code easier to read and debug.
```

```python
# ---------------------------
# GLOBAL RANDOM SEED
# ---------------------------

RANDOM_SEED = 42                  # Fixed seed so results are reproducible across runs.
random.seed(RANDOM_SEED)          # Apply the fixed seed to Python's random module.


# ---------------------------
# TIME SETTINGS
# ---------------------------

DAYS_PER_YEAR = 365               # We define 1 year as 365 simulation days.
SIM_YEARS = 30                    # Example long-run horizon: 30 years.
SIM_TIME = SIM_YEARS * DAYS_PER_YEAR  # Total simulation horizon in days.
Start_Day = 0 # Jan 1st 2026
Num_Reps = 10 # How many replications of the simulation we need to run
```

```python
# ---------------------------
# PATIENT DATA STRUCTURE
# ---------------------------
#
# Each patient carries baseline attributes and a dynamic state.
# In a real project, these fields would likely be initialized
# from historical NYP data distributions or actual patient records.

@dataclass
class Patient:
    patient_id: int                 # Unique patient identifier.
    #Demographic
    age: int                        # Patient age in years at simulation start.
    race: str                       # Patient race.
    ethnicity: str                  # Patient ethnicity.

    #Health History
    smoker: bool                    # Smoking history indicator.
    alcohol: bool

    bmi: float                      # Body mass index.
    hiv_positive: bool              # HIV status indicator.
    has_cervix: bool                # does patient have cervix
    Family_History: bool            # Does family have cancer history
    Fragility Fractures: bool       # Does the patient have fragile bone structure
    blood_type: str                 # What type of blood does the patient have
    patient_type: str               # Patient type (outpatient, inpatient and ER)

    # Dynamic fields below can change during the simulation:
    active: bool = True             # Indicates whether the patient is currently alive/active in model.
    treated_count: int = 0          # Counts how many successful treatments the patient has received.
    untreated_count: int = 0        # Counts how many unsuccessful / untreated positive paths occurred.
    waiting_time_PCP: int = 0       # Counts how many days have the patient been waiting
    waiting_time_gi: int = 0
    waiting_time_specialist: int = 0
    waiting_time_ER: int = 0

    #Days until last test
    Days_X_Ray: int
    Days_colonoscopy: int
    Days_stool_test: int
    Days_Mammography: int
    Days_MRI: int
    Days_Cytology: int
    Days_CT: int
    Days_Papsm: int
    Days_HPV: int
```

```python
# ---------------------------
# POPULATION INFO
# ---------------------------

ENTRY_DATA = [
    "num_eligible_women",         # average number of women entering the NYP system on a daily basis
    "pcp",                        # % of women entered through primary care physician. (only outpatient and drop-in)
    "gynecologist",               #  % of women entered through gynecologist. (outpatient, inpatient, and drop-in)
    "self_sought",                # S % of womenelf-initiated entry into care through a specialist appt. ()
    "other"                       # Any other entry route.
]

# ---------------------------
# CAPACITY INFO
# ---------------------------

ENTRY_CAPACITY=  [
    "num_pcps",                   # Number of primary care physicians in NYP system available on a random day.
    "num_pcp_outpatient_appts",   # Number of appoinments
    "num_pcp_dropin_appts",
    "num_gis",          # Number of gynecologists.
    "num_gis_outpatient_appts",
    "num_gis_inpatient_appts",
    "num_gis_dropin_appts",
    "num_specialist",
    "num_specialist_outpatients",
    "num_specialist_inpatients",
    "num_specialist_dropins",
    "num_ERs",
    "num_dropins"
                 ]


# ---------------------------
# ENTRY TO TEST TRANSITION PROPORTIONS
# ---------------------------
#TRANSITION_PROPORTIONS


# ---------------------------
# SCREENING TYPES
# ---------------------------

SCREENING_TYPES = [
    "cervical",                   # Cervical cancer screening.
    "lung",                       # Lung cancer screening.
    "breast",                     # Breast cancer screening.
    "colorectal",                 # Colorectal cancer screening.
    "osteoporosis"                # Osteoporosis screening.
]


# ---------------------------
# ENTRY POINT TYPES
# ---------------------------

ENTRY_POINTS = [
    "pcp",                        # Entered through primary care physician.
    "gynecologist",               # Entered through gynecologist.
    "self_sought",                # Self-initiated entry into care.
    "other"                       # Any other entry route.
]


# ---------------------------
# SCREENING OPTIONS BY CANCER
# ---------------------------
#
# For each screening program, we define a list of possible test
# options. These are placeholders and can be made much richer later.

SCREENING_OPTIONS = {
    "cervical": ["hpv", "cytology", "co_test"],         # Example cervical screening options.
    "lung": ["ldct"],                                   # Example lung screening option: low-dose CT.
    "breast": ["mammogram", "mri", "ultrasound"],       # Example breast screening options.
    "colorectal": ["fit", "colonoscopy", "sigmoidoscopy"], # Example colorectal screening options.
    "osteoporosis": ["dexa"]                            # Example osteoporosis screening option.
}





# ---------------------------
# HELPER FUNCTIONS:
# GENERATE BASELINE PATIENTS
# ---------------------------
#
# These are placeholder generators. In the real model, these
# would be estimated from the past 10 years of NYP female patient data.

def generate_entry_point() -> str:
    """Randomly assign an entry point using placeholder probabilities."""
    # We use weighted random selection to mimic different ways patients enter care.
    return random.choices(
        ENTRY_POINTS,               # Candidate entry points.
        weights=[0.45, 0.25, 0.20, 0.10],  # Placeholder probabilities.
        k=1                         # Choose one entry point.
    )[0]                           # Extract the chosen value from the returned list.


def generate_patient(patient_id: int) -> Patient:
    """Create one patient with placeholder health-history attributes."""
    age = random.randint(21, 80)   # Assign an age between 21 and 80.
    smoker = random.random() < 0.25  # 25% placeholder probability of smoking history.
    bmi = round(random.uniform(18.0, 40.0), 1)  # Random BMI between 18 and 40.
    systolic_bp = random.randint(100, 170)      # Placeholder systolic blood pressure.
    diastolic_bp = random.randint(60, 100)      # Placeholder diastolic blood pressure.
    hiv_positive = random.random() < 0.03       # 3% placeholder HIV prevalence.
    entry_point = generate_entry_point()         # Generate the patient’s entry point.

    # Return the complete Patient object.
    return Patient(
        patient_id=patient_id,      # Store patient ID.
        age=age,                    # Store age.
        smoker=smoker,              # Store smoking history.
        bmi=bmi,                    # Store BMI.
        systolic_bp=systolic_bp,    # Store systolic BP.
        diastolic_bp=diastolic_bp,  # Store diastolic BP.
        hiv_positive=hiv_positive,  # Store HIV status.
        entry_point=entry_point     # Store entry point.
    )


# ---------------------------
# HELPER FUNCTIONS:
# ELIGIBILITY RULES
# ---------------------------
#
# These rules are simplified placeholders. In the actual research
# model, they should reflect clinical guidelines, historical data,
# and the exact definition agreed upon with your professor.

def eligible_for_cervical(patient: Patient) -> bool:
    """Simplified eligibility rule for cervical screening."""
    return 21 <= patient.age <= 65  # Placeholder age-based rule.


def eligible_for_lung(patient: Patient) -> bool:
    """Simplified eligibility rule for lung screening."""
    return patient.age >= 50 and patient.smoker  # Placeholder smoking-based rule.


def eligible_for_breast(patient: Patient) -> bool:
    """Simplified eligibility rule for breast screening."""
    return patient.age >= 40  # Placeholder age-based rule.


def eligible_for_colorectal(patient: Patient) -> bool:
    """Simplified eligibility rule for colorectal screening."""
    return patient.age >= 45  # Placeholder age-based rule.


def eligible_for_osteoporosis(patient: Patient) -> bool:
    """Simplified eligibility rule for osteoporosis screening."""
    return patient.age >= 65 or patient.bmi < 19.0  # Placeholder age / low-BMI rule.


def get_eligible_screenings(patient: Patient) -> List[str]:
    """Return all screening programs the patient is currently eligible for."""
    eligible = []                                  # Start with an empty list.

    if eligible_for_cervical(patient):             # Check cervical eligibility.
        eligible.append("cervical")                # Add cervical if eligible.

    if eligible_for_lung(patient):                 # Check lung eligibility.
        eligible.append("lung")                    # Add lung if eligible.

    if eligible_for_breast(patient):               # Check breast eligibility.
        eligible.append("breast")                  # Add breast if eligible.

    if eligible_for_colorectal(patient):           # Check colorectal eligibility.
        eligible.append("colorectal")              # Add colorectal if eligible.

    if eligible_for_osteoporosis(patient):         # Check osteoporosis eligibility.
        eligible.append("osteoporosis")            # Add osteoporosis if eligible.

    return eligible                                # Return the final list.


# ---------------------------
# HELPER FUNCTIONS:
# SCREENING INTERVALS
# ---------------------------
#
# These values represent the number of days until the next due
# screening. In the real project, these intervals should be tied
# to guideline logic, prior results, risk category, and history.

def next_screening_interval_days(screening_type: str) -> int:
    """Return the number of days until the next routine screening."""
    if screening_type == "cervical":
        return 3 * DAYS_PER_YEAR                   # Placeholder: every 3 years.
    elif screening_type == "lung":
        return 1 * DAYS_PER_YEAR                   # Placeholder: annual.
    elif screening_type == "breast":
        return 1 * DAYS_PER_YEAR                   # Placeholder: annual.
    elif screening_type == "colorectal":
        return 2 * DAYS_PER_YEAR                   # Placeholder: simplified interval.
    elif screening_type == "osteoporosis":
        return 2 * DAYS_PER_YEAR                   # Placeholder: simplified interval.
    else:
        return 1 * DAYS_PER_YEAR                   # Safe default if type is unknown.


# ---------------------------
# HELPER FUNCTIONS:
# CHOOSE SCREENING OPTION
# ---------------------------

def choose_screening_option(screening_type: str) -> str:
    """Choose one screening modality from the available options."""
    options = SCREENING_OPTIONS[screening_type]    # Get the options for that cancer type.
    return random.choice(options)                  # Randomly choose one option for now.


# ---------------------------
# HELPER FUNCTIONS:
# SCREENING RESULT PROBABILITIES
# ---------------------------
#
# These are intentionally simple placeholders. In reality, the
# probability of a positive result should depend on patient risk
# profile, age, prior history, and maybe time since last screen.

def positive_screen_probability(patient: Patient, screening_type: str) -> float:
    """Return a placeholder probability of a positive screening result."""
    base_prob = 0.05                               # Start with a baseline positive rate.

    if screening_type == "lung" and patient.smoker:
        base_prob += 0.08                          # Smokers get higher positive probability for lung path.

    if screening_type == "breast" and patient.age >= 50:
        base_prob += 0.02                          # Slightly increased placeholder risk for older age.

    if screening_type == "osteoporosis" and patient.bmi < 19.0:
        base_prob += 0.05                          # Lower BMI increases osteoporosis-related risk.

    if patient.hiv_positive and screening_type == "cervical":
        base_prob += 0.04                          # Placeholder higher cervical risk for HIV-positive patients.

    return min(base_prob, 0.80)                    # Cap probability at 0.80 for numerical safety.


def is_positive_screen(patient: Patient, screening_type: str) -> bool:
    """Draw a positive vs negative screening outcome."""
    p = positive_screen_probability(patient, screening_type)  # Get patient-specific positive probability.
    return random.random() < p                  # Return True if random draw falls below probability.


# ---------------------------
# HELPER FUNCTIONS:
# TREATMENT SUCCESS
# ---------------------------
#
# Again, these are placeholders. A real version should reflect
# stage, treatment pathway, adherence, and outcome data.

def treatment_success_probability(patient: Patient, screening_type: str) -> float:
    """Return a placeholder probability that treatment succeeds."""
    base_prob = 0.75                             # Start with a general treatment success rate.

    if patient.smoker and screening_type == "lung":
        base_prob -= 0.10                        # Smoking reduces success in this placeholder logic.

    if patient.hiv_positive and screening_type == "cervical":
        base_prob -= 0.05                        # HIV reduces success slightly in placeholder logic.

    if patient.systolic_bp > 150:
        base_prob -= 0.03                        # Poor blood pressure control lowers success slightly.

    return max(0.20, min(base_prob, 0.95))       # Keep the probability in a reasonable range.


def treatment_is_successful(patient: Patient, screening_type: str) -> bool:
    """Draw whether treatment is successful."""
    p = treatment_success_probability(patient, screening_type)  # Get treatment success probability.
    return random.random() < p                  # Draw success/failure.


# ---------------------------
# HELPER FUNCTIONS:
# SERVICE DURATIONS
# ---------------------------
#
# All times are in DAYS because the simulation clock uses days.
# Some activities are shorter than a day, so we model them as
# fractional days.

def intake_duration_days() -> float:
    """Duration of initial intake/triage."""
    return random.uniform(0.05, 0.20)           # Roughly 1.2 to 4.8 hours.


def screening_duration_days(screening_type: str, option: str) -> float:
    """Duration of the screening visit itself."""
    # These are placeholders only.
    if screening_type == "cervical":
        return random.uniform(0.05, 0.10)       # Short outpatient visit.
    elif screening_type == "lung":
        return random.uniform(0.10, 0.20)       # Imaging visit.
    elif screening_type == "breast":
        return random.uniform(0.08, 0.20)       # Mammography / imaging.
    elif screening_type == "colorectal":
        return random.uniform(0.15, 0.50)       # Colonoscopy path may be longer.
    elif screening_type == "osteoporosis":
        return random.uniform(0.05, 0.10)       # DEXA tends to be short.
    else:
        return random.uniform(0.05, 0.20)       # Safe fallback.


def treatment_duration_days(screening_type: str) -> float:
    """Duration of treatment episode or treatment access wait placeholder."""
    # This is not true clinical treatment duration. It is just a placeholder
    # for the amount of simulated resource time a treatment slot consumes.
    if screening_type == "cervical":
        return random.uniform(10, 60)           # Placeholder treatment episode length.
    elif screening_type == "lung":
        return random.uniform(20, 120)          # Placeholder treatment episode length.
    elif screening_type == "breast":
        return random.uniform(20, 90)           # Placeholder treatment episode length.
    elif screening_type == "colorectal":
        return random.uniform(15, 100)          # Placeholder treatment episode length.
    elif screening_type == "osteoporosis":
        return random.uniform(30, 180)          # Placeholder long management horizon.
    else:
        return random.uniform(10, 60)           # Safe fallback.


def post_treatment_reentry_delay_days() -> int:
    """Delay before a treated patient re-enters routine long-run screening."""
    return random.randint(180, 2 * DAYS_PER_YEAR)  # Re-enter after 6 months to 2 years.


def post_negative_reentry_delay_days(screening_type: str) -> int:
    """Delay before next routine screen after a negative result."""
    return next_screening_interval_days(screening_type)  # Negative screens return to standard interval.


# ---------------------------
# METRICS CONTAINER
# ---------------------------
#
# This dictionary collects outputs for later analysis.

def initialize_metrics() -> Dict:
    """Create a metrics dictionary."""
    return {
        "screenings_completed": 0,              # Total number of completed screenings.
        "positive_screens": 0,                  # Number of positive screening results.
        "negative_screens": 0,                  # Number of negative screening results.
        "treatments_started": 0,                # Number of treatment episodes started.
        "treated_successfully": 0,              # Number of successful treatments.
        "untreated_or_failed": 0,               # Number of unsuccessful / untreated paths.
        "total_screening_wait": 0.0,            # Sum of queue waiting times before screening.
        "total_treatment_wait": 0.0,            # Sum of queue waiting times before treatment.
        "patient_cycles": 0,                    # Counts repeated patient screening cycles.
        "screenings_by_type": {s: 0 for s in SCREENING_TYPES},   # Count screenings by type.
        "positives_by_type": {s: 0 for s in SCREENING_TYPES},    # Count positives by type.
        "treated_by_type": {s: 0 for s in SCREENING_TYPES},      # Count successful treatments by type.
        "patient_history": {}                   # Optional per-patient event history for later inspection.
    }


# ---------------------------
# RESOURCE BUILDER
# ---------------------------
#
# We create one intake resource, one screening resource per cancer,
# and one treatment resource per cancer. Capacities are placeholders.

def build_resources(env: simpy.Environment) -> Dict:
    """Create all capacity-constrained resources used in the model."""
    resources = {}

    resources["intake"] = simpy.Resource(env, capacity=10)  # Intake / scheduling / triage capacity.

    # One screening resource per program.
    resources["screening"] = {
        "cervical": simpy.Resource(env, capacity=8),        # Placeholder cervical screening capacity.
        "lung": simpy.Resource(env, capacity=4),            # Placeholder lung screening capacity.
        "breast": simpy.Resource(env, capacity=10),         # Placeholder breast screening capacity.
        "colorectal": simpy.Resource(env, capacity=6),      # Placeholder colorectal screening capacity.
        "osteoporosis": simpy.Resource(env, capacity=5)     # Placeholder osteoporosis screening capacity.
    }

    # One treatment resource per program.
    resources["treatment"] = {
        "cervical": simpy.Resource(env, capacity=5),        # Placeholder cervical treatment capacity.
        "lung": simpy.Resource(env, capacity=3),            # Placeholder lung treatment capacity.
        "breast": simpy.Resource(env, capacity=6),          # Placeholder breast treatment capacity.
        "colorectal": simpy.Resource(env, capacity=4),      # Placeholder colorectal treatment capacity.
        "osteoporosis": simpy.Resource(env, capacity=5)     # Placeholder osteoporosis management capacity.
    }

    return resources                                        # Return all built resources.


# ---------------------------
# HISTORY LOGGING
# ---------------------------
#
# This helper stores a simple event history for each patient.

def log_event(metrics: Dict, patient: Patient, time_now: float, message: str) -> None:
    """Append a time-stamped event message to the patient history."""
    # Create the patient's log if it does not exist yet.
    if patient.patient_id not in metrics["patient_history"]:
        metrics["patient_history"][patient.patient_id] = []

    # Append the event message.
    metrics["patient_history"][patient.patient_id].append((time_now, message))


# ---------------------------
# PATIENT PROCESS
# ---------------------------
#
# This is the main SimPy process. Each patient continuously cycles
# through the healthcare system over the long horizon.
#
# Key idea:
# The patient does NOT disappear after one screening.
# Instead, they can re-enter later and continue screening over time.

def patient_process(env: simpy.Environment, patient: Patient, resources: Dict, metrics: Dict):
    """SimPy process representing one patient over many years."""
    # We start with an initial short delay before first contact with the system.
    initial_delay = random.randint(0, 90)                   # First system entry occurs sometime in first 90 days.
    yield env.timeout(initial_delay)                        # Wait until that first contact time.
    log_event(metrics, patient, env.now, "Initial entry into system")  # Log initial system entry.

    # This loop allows the same patient to repeatedly come back over decades.
    while patient.active:
        metrics["patient_cycles"] += 1                      # Count one new patient cycle.

        # ---------------------------------------
        # STEP 1: INTAKE / ENTRY POINT CONTACT
        # ---------------------------------------
        #
        # The patient first uses the intake resource. This can represent
        # scheduling, triage, clinic routing, or entry-point processing.

        intake_queue_start = env.now                        # Record when the patient starts waiting for intake.

        with resources["intake"].request() as req:          # Request access to intake capacity.
            yield req                                       # Wait until intake capacity is available.
            intake_wait = env.now - intake_queue_start      # Compute intake wait.
            log_event(metrics, patient, env.now, f"Entered via {patient.entry_point}; intake wait = {intake_wait:.2f} days")  # Log entry.
            yield env.timeout(intake_duration_days())       # Spend time in intake.

        # ---------------------------------------
        # STEP 2: DETERMINE ELIGIBLE SCREENINGS
        # ---------------------------------------
        #
        # Based on age and risk profile, find all screening programs
        # for which the patient is currently eligible.

        eligible_screenings = get_eligible_screenings(patient)  # Get all currently eligible screening types.
        log_event(metrics, patient, env.now, f"Eligible screenings: {eligible_screenings}")  # Log eligibility.

        # If the patient is currently not eligible for any screening,
        # we age the patient forward and try again later.
        if not eligible_screenings:                         # Check whether no screening is currently available.
            wait_until_next_review = 180                    # Wait 180 days before reassessing.
            yield env.timeout(wait_until_next_review)       # Advance simulation time.
            patient.age += wait_until_next_review // DAYS_PER_YEAR  # Age the patient approximately if enough time passes.
            continue                                        # Restart the loop for this patient.

        # ---------------------------------------
        # STEP 3: CHOOSE ONE SCREENING TO EXECUTE
        # ---------------------------------------
        #
        # For a first framework, we choose one eligible screening in
        # the current cycle. Later you can make this more realistic by:
        #   - prioritizing certain screenings,
        #   - running multiple screenings in same cycle,
        #   - following historical utilization patterns.

        screening_type = random.choice(eligible_screenings) # Choose one eligible screening path for this cycle.
        screening_option = choose_screening_option(screening_type)  # Choose the modality/test option.
        log_event(metrics, patient, env.now, f"Assigned to {screening_type} screening using {screening_option}")  # Log choice.

        # ---------------------------------------
        # STEP 4: SCREENING QUEUE AND SERVICE
        # ---------------------------------------
        #
        # Patient requests the appropriate screening resource and waits
        # in line if the service is busy.

        screening_queue_start = env.now                     # Record queue start time for screening.

        with resources["screening"][screening_type].request() as req:  # Request the screening resource for this cancer type.
            yield req                                       # Wait until screening slot becomes available.
            screening_wait = env.now - screening_queue_start  # Measure queue wait.
            metrics["total_screening_wait"] += screening_wait # Add to total screening wait metric.
            log_event(metrics, patient, env.now, f"Started {screening_type} screening after waiting {screening_wait:.2f} days")  # Log screening start.
            yield env.timeout(screening_duration_days(screening_type, screening_option))  # Spend time in the screening service.
            log_event(metrics, patient, env.now, f"Completed {screening_type} screening ({screening_option})")  # Log screening completion.

        # Update screening metrics after service is completed.
        metrics["screenings_completed"] += 1                # Increase total completed screenings.
        metrics["screenings_by_type"][screening_type] += 1  # Increase count for this screening type.

        # ---------------------------------------
        # STEP 5: SCREENING RESULT
        # ---------------------------------------
        #
        # The patient receives either a negative or positive result.

        positive = is_positive_screen(patient, screening_type)  # Draw screen result.

        if not positive:                                    # If the result is negative:
            metrics["negative_screens"] += 1                # Count a negative screen.
            log_event(metrics, patient, env.now, f"Negative {screening_type} screen")  # Log negative result.

            # Wait until the next recommended screening interval.
            delay = post_negative_reentry_delay_days(screening_type)  # Compute next routine re-entry delay.
            yield env.timeout(delay)                        # Advance time until next routine return.
            patient.age += delay // DAYS_PER_YEAR           # Update age approximately.
            continue                                        # Return to the beginning of the patient loop.

        # If the result is positive, proceed to treatment.
        metrics["positive_screens"] += 1                    # Count a positive screen.
        metrics["positives_by_type"][screening_type] += 1   # Count positive by type.
        log_event(metrics, patient, env.now, f"Positive {screening_type} screen")  # Log positive result.

        # ---------------------------------------
        # STEP 6: TREATMENT QUEUE AND SERVICE
        # ---------------------------------------
        #
        # Positive patients request treatment capacity for the relevant
        # disease area.

        treatment_queue_start = env.now                     # Record treatment queue start.

        with resources["treatment"][screening_type].request() as req:  # Request disease-specific treatment capacity.
            yield req                                       # Wait until treatment capacity becomes available.
            treatment_wait = env.now - treatment_queue_start  # Compute treatment wait.
            metrics["total_treatment_wait"] += treatment_wait # Add to treatment wait metric.
            metrics["treatments_started"] += 1              # Count a treatment start.
            log_event(metrics, patient, env.now, f"Started treatment for {screening_type} after waiting {treatment_wait:.2f} days")  # Log treatment start.
            yield env.timeout(treatment_duration_days(screening_type))  # Spend time in treatment episode.

        # ---------------------------------------
        # STEP 7: TREATMENT OUTCOME
        # ---------------------------------------
        #
        # Treatment is either successful or unsuccessful. In your project,
        # “untreated” may also include a branch for non-completion or
        # unresolved disease. Here we combine them into one placeholder.

        success = treatment_is_successful(patient, screening_type)  # Draw treatment success.

        if success:                                         # If treatment succeeds:
            metrics["treated_successfully"] += 1            # Count successful treatment.
            metrics["treated_by_type"][screening_type] += 1 # Count successful treatment by cancer type.
            patient.treated_count += 1                      # Update the patient’s personal count.
            log_event(metrics, patient, env.now, f"Successfully treated for {screening_type}")  # Log success.

            # After successful treatment, the patient eventually re-enters
            # for future screening over the long horizon.
            delay = post_treatment_reentry_delay_days()     # Determine how long until re-entry.
            yield env.timeout(delay)                        # Wait until the patient returns.
            patient.age += delay // DAYS_PER_YEAR           # Update age approximately.
            continue                                        # Start the next cycle.

        else:                                               # If treatment fails or patient remains untreated:
            metrics["untreated_or_failed"] += 1             # Count unsuccessful / untreated outcome.
            patient.untreated_count += 1                    # Update patient’s personal count.
            log_event(metrics, patient, env.now, f"Unsuccessful treatment / untreated for {screening_type}")  # Log failure.

            # Even after unsuccessful treatment, the patient may stay in the
            # system and re-enter later; this is a modeling choice.
            delay = random.randint(90, 365)                # Reassess after 3 to 12 months.
            yield env.timeout(delay)                       # Wait before next system interaction.
            patient.age += delay // DAYS_PER_YEAR          # Update age approximately.
            continue                                        # Continue patient lifecycle.


# ---------------------------
# PATIENT COHORT CREATION
# ---------------------------
#
# This function generates the initial panel of patients that the
# simulation will track over time.

def create_initial_cohort(n_patients: int) -> List[Patient]:
    """Create an initial cohort of patients."""
    cohort = []                                            # Start with an empty list.

    for pid in range(1, n_patients + 1):                   # Loop over patient IDs.
        cohort.append(generate_patient(pid))               # Generate and add each patient.

    return cohort                                          # Return the completed cohort.


# ---------------------------
# RUN ONE SIMULATION
# ---------------------------

def run_simulation(n_patients: int = 1000, sim_time: int = SIM_TIME) -> Dict:
    """Run one long-horizon simulation and return collected metrics."""
    env = simpy.Environment()                              # Create the SimPy environment (simulation clock).
    metrics = initialize_metrics()                         # Create metrics storage.
    resources = build_resources(env)                       # Build all resources.
    cohort = create_initial_cohort(n_patients)            # Generate initial patient cohort.

    # Start one SimPy process for each patient.
    for patient in cohort:                                 # Loop through every patient in the cohort.
        env.process(patient_process(env, patient, resources, metrics))  # Launch the patient lifecycle process.

    # Run the simulation until the chosen horizon in days.
    env.run(until=sim_time)                                # Advance simulation until time horizon is reached.

    return metrics                                         # Return all collected metrics.


# ---------------------------
# SUMMARY PRINTER
# ---------------------------
#
# This helper prints a readable summary after a run.

def print_summary(metrics: Dict) -> None:
    """Print a simple summary of simulation results."""
    print("=" * 60)                                        # Print a visual separator.
    print("SIMULATION SUMMARY")                            # Print title.
    print("=" * 60)                                        # Print separator again.

    print(f"Total screenings completed: {metrics['screenings_completed']}")    # Total screening count.
    print(f"Positive screens:           {metrics['positive_screens']}")         # Positive count.
    print(f"Negative screens:           {metrics['negative_screens']}")         # Negative count.
    print(f"Treatments started:         {metrics['treatments_started']}")       # Treatment count.
    print(f"Successfully treated:       {metrics['treated_successfully']}")     # Successful treatment count.
    print(f"Unsuccessful / untreated:   {metrics['untreated_or_failed']}")      # Unsuccessful treatment count.
    print(f"Patient cycles completed:   {metrics['patient_cycles']}")           # Number of patient cycles.

    # Compute average waits only if there were relevant events.
    avg_screen_wait = (
        metrics["total_screening_wait"] / metrics["screenings_completed"]
        if metrics["screenings_completed"] > 0 else 0.0
    )                                                      # Average screening wait.

    avg_treat_wait = (
        metrics["total_treatment_wait"] / metrics["treatments_started"]
        if metrics["treatments_started"] > 0 else 0.0
    )                                                      # Average treatment wait.

    print(f"Average screening wait:     {avg_screen_wait:.2f} days")           # Print screening wait.
    print(f"Average treatment wait:     {avg_treat_wait:.2f} days")            # Print treatment wait.

    print("\nScreenings by type:")                         # Section header.
    for screening_type, count in metrics["screenings_by_type"].items():  # Loop through screening counts.
        print(f"  {screening_type:12s}: {count}")          # Print per-type screening count.

    print("\nPositive screens by type:")                   # Section header.
    for screening_type, count in metrics["positives_by_type"].items():  # Loop through positive counts.
        print(f"  {screening_type:12s}: {count}")          # Print per-type positive count.

    print("\nSuccessful treatments by type:")              # Section header.
    for screening_type, count in metrics["treated_by_type"].items():  # Loop through treatment counts.
        print(f"  {screening_type:12s}: {count}")          # Print per-type successful treatment count.

    print("=" * 60)                                        # Final separator.


# ---------------------------
# MAIN BLOCK
# ---------------------------
#
# This block only runs if the file is executed directly.

if __name__ == "__main__":                                 # Standard Python entry point check.
    results = run_simulation(n_patients=1000, sim_time=SIM_TIME)  # Run the simulation.
    print_summary(results)                                 # Print a summary of outputs.

    # Optional: print a short example patient history for one patient.
    sample_patient_id = 1                                  # Pick one patient to inspect.
    if sample_patient_id in results["patient_history"]:    # Check that the patient exists in the history log.
        print(f"\nSample event history for patient {sample_patient_id}:")  # Print header.
        for event_time, message in results["patient_history"][sample_patient_id][:15]:  # Show first 15 events.
            print(f"Day {event_time:.2f}: {message}")      # Print the event.
```
